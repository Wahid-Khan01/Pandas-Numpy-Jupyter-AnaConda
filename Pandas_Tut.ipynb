```python
dict1 = {
    'a':['abc',np.nan,pd.NaT,'def', 'proof'],
    'b':[1,2,3,4,5],
    'c':['proof',pd.NaT,pd.NaT,np.nan, 'proof']
}
```

```python
drfr = pd.DataFrame(dict1)
```

```python
drfr
```

```python
drfr.drop_duplicates(subset=['c'])  # yeh a column me se duplicate o hata k by default sirf first waaley ko rakhega
```

```python
drfr.drop_duplicates(subset=['c'], keep='first')  # duplicate ko hata k pehle wala rakhega jo ki bydefault hota hai
drfr.drop_duplicates(subset=['c'], keep='last')  # first duplicate ko hata ke last waaley ko rakhunga
drfr.drop_duplicates(subset=['c'], keep='False') #yeh sabhi duplicate ko remove kardega or duplicates me se koi bhi nahi rakhega
```
